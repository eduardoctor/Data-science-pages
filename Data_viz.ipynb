### Using AutoViz
This will create a visualization in-line of the main variables

```python
from autoviz.AutoViz_Class import AutoViz_Class
import pandas as pd
AV = AutoViz_Class()

df_av = AV.AutoViz('SearchResults_May2020.csv')
```

### Using Sweetviz
This will create a more detailed report on an HTML file on a different tab

```python
df = pd.read_csv('SearchResults_May2020.csv')
import sweetviz as sv
my_report = sv.analyze(df)
my_report.show_html()
```
